```python
#selecionando o IDH por pais

dados3= pd.read_excel('IDH.xlsx')
dados3.loc[:,['geo',2015]]

#renomeando as colunas

dados3=dados3.loc[ :,["geo", 2015]]
dados3.rename(columns={2015:"IDH","geo":"País"},inplace=True)
dados3.set_index("País",inplace=True)

d30 = dados3.head(20)
d30


```

```python
# juntando as colunas em um único dataframe

junto=[dados3,dados1,dados2]
first=junto.pop(0)

df=first.join(junto,how="inner")

df.head(20)
```

```python
#criando gráfico IDH X razão entre mulheres e homens nas escolas

sns.regplot(df["IDH"],df["Mulheres/Homens nas escolas"], color=('green'))
plt.grid()
plt.title('Mulheres/Homens nas Escolas X IDH')
plt.grid(True)
plt.show()

# pegando a correlação entre IDH e a razão entre mulheres e homens nas escolas
corr1 = dados1['Mulheres/Homens nas escolas'].corr(dados3['IDH'])
print('Correlação entre varíaveis:'+str(corr1))
```

A partir do grafico acima, conclui-se que, na maioria dos casos, os países com IDH mais elevados tendem a ter uma razção mais próxima de 1. Isso mostra que o número de homens e mulheres nas escolas é muito próximo. Nota-se tambem que o caso em que o IDH é o mais baixo, a razão é menor que 1. Com isso, conclui-se que, neste país, o número de mulheres que frequenta escolas é maior que o de homens. Entretanto, nota-se que não necessariamente países com IDH maiores têm uma razão mais proóxima de 1. A correlação entra as variáveis é muito baixa, o que confirma o que foi dito.
